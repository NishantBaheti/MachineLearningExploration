## Box Plot

```python
df_setosa = df[df['species'] == "setosa"]
df_versicolor = df[df['species'] == "versicolor"]
df_virginica = df[df['species'] == "virginica"]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20,5))
sns.boxplot(data=df_setosa, ax=ax[0])
ax[0].set_title('setosa')

sns.boxplot(data=df_versicolor, ax=ax[1])
ax[1].set_title('versicolor')

sns.boxplot(data=df_virginica, ax=ax[2])
ax[2].set_title('virginica')

plt.show()
```

So what did it tell

1. data ranges and scales are different. Possibly a standardization would be necessary.
2. not many outliers. 

## Pairplot to check relations and distribution

```python
sns.set(font_scale=1.2)
sns.pairplot(df,hue="species", height=3, markers='o')
```

`observations`

- in scatter plots it is visible that setosa is having properties much different than other two and cluster can be created.
- verginica and versicolor is showing some overlap on the data points but still two clusters can be maintained here.

### cross product

### magnitude $||\vec{A}||$ 

```python
(
    x, 
    np.sqrt(np.sum(x**2)), 
    np.linalg.norm(x)
)
```

### project $\vec{A}$ to $\vec{B}$

$$
proj_\vec{A}(\vec{B}) = (\vec{A}.\vec{B}).\frac{\vec{B}}{||\vec{B}||^2}
$$

```python
def projection(A, B):
    return (A @ B.T) * (B / np.linalg.norm(B)**2)
```

```python
A = np.array([[1,7]]) 
B = np.array([[-4,2]])

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(2,2,(1,2))

ax.scatter(A[:,[0]],A[:,[1]],s=70,label="A")
ax.scatter(B[:,[0]],B[:,[1]],s=70,label="B")


for i in A:
    ax.plot([0,i[0]],[0,i[1]],alpha=0.7,label='A')
for i in B:
    ax.plot([0,i[0]],[0,i[1]],alpha=0.7,label='B')

ax.legend(loc='best')


ax = fig.add_subplot(2,2,(3,4))

ax.scatter(A[:,[0]],A[:,[1]],s=70,label='A')
ax.scatter(B[:,[0]],B[:,[1]],s=70,label='B')

for i in A:
    ax.plot([0,i[0]],[0,i[1]],alpha=0.7,label='A')
for i in B:
    ax.plot([0,i[0]],[0,i[1]],alpha=0.7,label='B')
    
projection = projection(A,B)

ax.scatter(projection[:,[0]],projection[:,[1]],s=70,label="projection",c='black')
for i in projection:
    ax.plot([0,i[0]],[0,i[1]],c='black',alpha=0.7,label="projection",ls='--')

ax.legend(loc='best')
plt.show()
```
